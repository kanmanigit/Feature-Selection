```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as py
```

```python
df=pd.read_csv("prep.csv")
df
```

```python
df1=pd.get_dummies(df,drop_first=True,dtype=int)
df1
```

```python
df1.columns
```

```python
indep=df1.drop('classification_yes',axis=1)
indep.columns
dep=df1["classification_yes"]
```

```python
indep.shape
#dep
```

```python
# Split the data into training and test datasets
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain,yTest = train_test_split(indep, dep, test_size=0.30, random_state=0)
```

```python
# StandardScalar obj
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
xtrain_Scaled = SS.fit_transform(xTrain)
xtest_Scaled = SS.fit_transform(xTest)
```

```python
# without feature selection check accuracy with Ramdam forest

from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=50, max_depth=None,min_samples_split=2, random_state=100)
rf_classifier.fit(xtrain_Scaled, yTrain)

y_pred = rf_classifier.predict(xtest_Scaled)

# Accuracy
from sklearn.metrics import accuracy_score
accuracy_score =accuracy_score(yTest,y_pred)
print("Accuracy score= ", accuracy_score)
```

```python
# get the Importance of the resulting features...

feature_importance = rf_classifier.feature_importances_
feature_importance
```

```python
# create the Dataframe
final_df1 = pd.DataFrame({'Features':indep.columns, 'Importance Scores':feature_importance})
final_df1
```

```python
# sort in ascending order to better visualization

final_df1 = final_df1.sort_values(by='Importance Scores',ascending=False)
final_df1
```

```python
# plot the feature importance in Bars...

plot.bar( final_df1["Features"], final_df1["Importance Scores"])
plot.title( "Feature Importance in CKD" )
plot.xlabel( "Features" )
plot.ylabel( "Importance Scores" )

# Rotate x-axis labels for better readability
plot.xticks(rotation=50)
plot.show
```

```python
# with feature selection check accuracy with Ramdom Forest

rfe_classfier =RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)

from sklearn.feature_selection import RFE

sel_rfe_classifier = RFE(estimator=rfe_classfier, n_features_to_select=7 )

rfe_fit = sel_rfe_classifier.fit(xtrain_Scaled,yTrain)

xtrain_new = rfe_fit.transform(xtrain_Scaled)




selected_features=indep.columns[rfe_fit.get_support()]
selected_features
```

```python
ypred =sel_rfe_classifier.predict(xtest_Scaled)

from sklearn.metrics import accuracy_score
acc=accuracy_score(yTest, ypred)
print ("accuracy score after RFE ", acc)
```

```python
# feature importance

# Now, fit a new RandomForestClassifier on the selected features

fitted_rf_model = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
fitted_rf_model.fit(xtrain_new, yTrain)

feature_importance=fitted_rf_model.feature_importances_
feature_importance
```

```python
score = pd.DataFrame( feature_importance, columns=["Score"] )
col = pd.DataFrame( selected_features, columns=["features"] )

#impor
#x_col
fdf =pd.concat([col,score],axis=1)
fdf
```

```python
fdf_1=fdf.sort_values(by="Score",ascending=False)

plot.bar( fdf_1["features"], fdf_1["Score"])
plot.title("Feature Importance")
plot.xlabel( "Features" )
plot.ylabel( "Importance Scores" )
```
