```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as py
```

```python
df=pd.read_csv("prep.csv")
df
```

```python
df1=pd.get_dummies(df,drop_first=True,dtype=int)
df1
```

```python
df1.columns
```

```python
indep=df1.drop('classification_yes',axis=1)
indep.columns
dep=df1["classification_yes"]
```

```python
indep.shape
#dep
```

```python
# Split the data into training and test datasets
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain,yTest = train_test_split(indep, dep, test_size=0.30, random_state=0)
```

```python
# StandardScalar obj
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
xtrain_Scaled = SS.fit_transform(xTrain)
xtest_Scaled = SS.fit_transform(xTest)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=10, max_depth=None,min_samples_split=2, random_state=0)
rf_classifier.fit(xtrain_Scaled, yTrain)
```

```python
feature_importance = rf_classifier.feature_importances_
feature_importance
```

```python
df1 = pd.DataFrame({'Features':indep.columns, 'Importance Scores':feature_importance})
df1
```

```python
df2 = df1.sort_values(by='Importance Scores',ascending=False)
df2
```

```python
plot.bar( df2["Features"], df2["Importance Scores"])
plot.title( "Feature Importance in CKD" )
plot.xlabel( "Features" )
plot.ylabel( "Importance Scores" )

# Rotate x-axis labels for better readability
plot.xticks(rotation=50)
plot.show
```
